```python
x1 = 3
x2 = 2
yactual = 17
```

```python
w1 = 1.1 # random values for weights
w2 = 0.7
```

```python
ypred = x1*w1 + x2*w2
print(ypred)
```

```python
error = (yactual-ypred)**2
error
```

```python
def forward_pass(x1, x2, w1, w2):
  ypred = x1 * w1 + x2 * w2
  return ypred
```

```python
learning_rate = 0.01
iterations = 20
# Lists to store history for plotting
errors_history = []
ypred_history = []
```

```python
for i in range(iterations):
  ypred = forward_pass(x1, x2, w1, w2)
  error = (yactual - ypred)**2

  # Gradients
  d_ypred = -2 * (yactual - ypred)
  dw1 = d_ypred * x1
  dw2 = d_ypred * x2

  # Update weights
  w1 = w1 - learning_rate * dw1
  w2 = w2 - learning_rate * dw2

  # Store values
  errors_history.append(error)
  ypred_history.append(ypred)
```

```python
print(f"After 20 Iterations:")
print(f"  w1: {w1:.4f}, w2: {w2:.4f}")
print(f"  ypred: {ypred:.4f}, error: {error:.4f}")
```

```python
import matplotlib.pyplot as plt

# Plotting the progression of error
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(iterations), errors_history)
plt.title('Error Progression Across Epochs')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.grid(True)

# Plotting the progression of ypred
plt.subplot(1, 2, 2)
plt.plot(range(iterations), ypred_history)
plt.axhline(y=yactual, color='r', linestyle=':', label=f'Actual Y ({yactual})')
plt.title('Predicted Y Progression Across Epochs')
plt.xlabel('Epoch')
plt.ylabel('Predicted Y (ypred)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
